```python
import numpy as np
import matplotlib.pyplot as plt
import math

def erlang_b(traffic_intensity, number_of_channels):
    """Calculate the blocking probability using Erlang B formula."""
    if traffic_intensity < 0 or number_of_channels < 1:
        raise ValueError("Invalid input values.")
    
    numerator = (traffic_intensity**number_of_channels) / math.factorial(number_of_channels)
    denominator = sum([(traffic_intensity**i) / math.factorial(i) for i in range(number_of_channels + 1)])
    
    return numerator / denominator

# Range of traffic intensity values
traffic_intensity_values = np.linspace(0, 50, 100)

# Calculate blocking probability for each traffic intensity value
blocking_probabilities = [erlang_b(traffic_intensity, 20) for traffic_intensity in traffic_intensity_values]

# Plotting the results
plt.plot(traffic_intensity_values, blocking_probabilities)
plt.xlabel('Traffic Intensity')
plt.ylabel('Blocking Probability')
plt.title('Erlang B Blocking Probability')
plt.grid(True)
plt.show()
```

```python
#複数チャネルを表示
# Path: erlang_b_multiple_channels.ipynb
import numpy as np
import matplotlib.pyplot as plt
import math

def erlang_b(traffic_intensity, number_of_channels):
    """Calculate the blocking probability using Erlang B formula."""
    if traffic_intensity < 0 or number_of_channels < 1:
        raise ValueError("Invalid input values.")
    
    numerator = (traffic_intensity**number_of_channels) / math.factorial(number_of_channels)
    denominator = sum([(traffic_intensity**i) / math.factorial(i) for i in range(number_of_channels + 1)])
    
    return numerator / denominator

# Range of traffic intensity values
traffic_intensity_values = np.linspace(0, 100, 100)

# Calculate blocking probability for each traffic intensity value
blocking_probabilities_10_channels = [erlang_b(traffic_intensity, 10) for traffic_intensity in traffic_intensity_values]
blocking_probabilities_20_channels = [erlang_b(traffic_intensity, 20) for traffic_intensity in traffic_intensity_values]
blocking_probabilities_30_channels = [erlang_b(traffic_intensity, 30) for traffic_intensity in traffic_intensity_values]
blocking_probabilities_40_channels = [erlang_b(traffic_intensity, 40) for traffic_intensity in traffic_intensity_values]
blocking_probabilities_50_channels = [erlang_b(traffic_intensity, 50) for traffic_intensity in traffic_intensity_values]

# Plotting the results
plt.plot(traffic_intensity_values, blocking_probabilities_10_channels, label='10 channels')
plt.plot(traffic_intensity_values, blocking_probabilities_20_channels, label='20 channels')
plt.plot(traffic_intensity_values, blocking_probabilities_30_channels, label='30 channels')
plt.plot(traffic_intensity_values, blocking_probabilities_40_channels, label='40 channels')
plt.plot(traffic_intensity_values, blocking_probabilities_50_channels, label='50 channels')
plt.xlabel('Traffic Intensity')
plt.ylabel('Blocking Probability')
plt.title('Erlang B Blocking Probability')
plt.grid(True)
plt.legend()
plt.show()
```
